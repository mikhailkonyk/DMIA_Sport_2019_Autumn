```python
df2 = df.copy()

asw_col = 'Times'

cols_to_del = []

#смотрим корреляцию с учителем:
def del_corr_xy(df, asw_col = None):
    global cols_to_del
    corr_y = df[list(set(list(df)) - set(asw_col))]\
        .corrwith(df[asw_col]).abs()\
                .sort_values(ascending = False)
    do_del = corr_y[abs(corr_y) < 0.01].index.tolist()
    cols_to_del += do_del
    return do_del

cols_to_del = del_corr_xy(df, asw_col = asw_col)

df = df.drop(cols_to_del, 1)
```

```python
cols_to_del = []

#смотрим корреляцию между собой:

def corr_matrix(df, asw_col, corr):
    features = list(set(list(df)) - set(asw_col))
    corr_x = df[features].corr().abs()
    corr_x = corr_x[(corr_x > corr) & (corr_x != 1)]
    for i in range(len(corr_x)):
        for j in range(len(list(corr_x))):
            if i > j:
                corr_x[list(corr_x)[j]].iloc[i] = np.nan
    return corr_x

def get_ft_to_dell(df, asw_col, corr):
    global cols_to_del
    corr_x = corr_matrix(df, asw_col, corr)
    corr_x = corr_x.fillna(0)
    to_del = []
    for i in list(corr_x)[1:]:
        corr_x = corr_matrix(df, asw_col, corr)
        if len(corr_x[i][corr_x[i] > 0]) > 0:
            to_del.append(i)
    cols_to_del += to_del
    return to_del

def dell_corr_x(df, asw_col, corr):
    to_del = get_ft_to_dell(df, asw_col, corr)
    i = len(to_del)
    while i != 0:
        cols_to_del.append(to_del[i-1])
        df = df.drop(to_del[i-1], 1)
        to_del = get_ft_to_dell(df, asw_col, corr)
        i = len(to_del)
    return df

df = dell_corr_x(df, asw_col = asw_col, corr = 0.95)

del cols_to_del

#corr= df.corr(method='pearson')
```

```python
X, y = df[list(set(list(df)) - {'Password', 'Password_list', 'Times'})].values,\
        df['Times'].values
X = np.abs(X)


X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle = True, test_size  = 0.15)


model = GradientBoostingRegressor()


params={'min_samples_split': 3, 'min_samples_leaf': 3, 
        'max_features': 'log2', 'loss': 'huber', 'learning_rate': 0.1, 'alpha': 0.95}

model.set_params(**params)
```

```python
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred = pd.Series(data = y_pred).apply(lambda x: (x < 1 and 1) or x).values
print(rlmse(y_test, y_pred))
```
